# What are the most common primary and secondary typings? How many Pokemon have either of those type? How many pokemon have both of those types?

```python
type1_counts = df['type1'].value_counts()
type2_counts = df['type2'].value_counts()

common_type1 = type1_counts.idxmax()
common_type2 = type2_counts.idxmax()

count_type1 = type1_counts[common_type1]
count_type2 = type2_counts[common_type2]

common_types = df[(df['type1'] == common_type1) & (df['type2'] == common_type2)]

count_common_types = len(common_types)

print("The most common primary type is", common_type1, "with", count_type1, "Pokemon having this type!")
print("The most common secondary type is", common_type2, "with", count_type2, "Pokemon having this type!")
print("There are", count_common_types, "pokemon with both", common_type1, "and", common_type2, "type!")
```

# Which legendary Pokemon has the highest base stat total? Which non-legendary Pokemon has the highest base stat total?

```python
legendary = df[df['legendary']]
non_legendary= df[~df['legendary']]


# Find the index of the legendary Pokémon with the highest total
high_leg_total = legendary['total'].idxmax()

# Find the index of the non-legendary Pokémon with the highest total
high_non_leg_total = non_legendary['total'].idxmax()

# Get the names and total stats of the legendary and non-legendary Pokémon with the highest total
name_high_leg = legendary.loc[high_leg_total, 'name']
total_high_leg = legendary.loc[high_leg_total, 'total']

name_high_non_leg = non_legendary.loc[high_non_leg_total, 'name']
total_high_non_leg = non_legendary.loc[high_non_leg_total, 'total']

print("The legendary pokemon with the highest base stat total is",name_high_leg,"with a base stat total of", total_high_leg,"!")
print("The normal pokemon with the highest base stat total is",name_high_non_leg,"with a base stat total of", total_high_non_leg,"!")
```
